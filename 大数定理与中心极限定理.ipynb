```python
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def getN_randNum(n,max_N):
    return [random.randint(1,max_N) for i in range(n)]
```

```python
getN_randNum(10,6)
```

```python
sumN = []
sumNSub3_5n = []
sumNDivN = []
for j in range(10):
    sumN.append([])
    sumNSub3_5n.append([])
    sumNDivN.append([])
    for k in range(1,501):
        sumTemp = float(sum(getN_randNum(k,6)))
        sumN[j].append(sumTemp)
        sumTempSub3_5n = sumTemp - 3.5*k
        sumNSub3_5n[j].append(sumTempSub3_5n)
        sumTempdivN = sumTemp/k
        sumNDivN[j].append(sumTempdivN)
```

```python
sumN = np.asarray(sumN).T
sumNSub3_5n = np.asarray(sumNSub3_5n).T
sumNDivN = np.asarray(sumNDivN).T
```

```python
n_100 = []
n_500 = []
for i in range(50):
    n_100.append(sum(getN_randNum(100,6)) - 3.5*100)
    n_500.append(sum(getN_randNum(500,6)) - 3.5*500)
```

```python
plt.figure(1)
plt.plot(n_100,"o");
plt.figure(2)
plt.plot(n_500,"o");
```

```python
plt.figure(1)
plt.plot(sumN);
plt.figure(2)
plt.plot(sumNSub3_5n[:,0],"o");
plt.figure(3)
plt.plot(sumNDivN);
```
